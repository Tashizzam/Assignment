**Definition:** Optical Character Recognition (OCR) is the process of converting different types of documents, such as scanned paper documents, PDFs, or images captured by a camera, into editable and searchable data.

```python
# Importing necessary libraries
from google.colab import files
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
```

```python
# Uploading the image file
uploaded = files.upload()
```

```python
# Assuming only one file is uploaded
filename = list(uploaded.keys())[0]
```

```python
# Display the image using matplotlib
img = Image.open(filename)
plt.imshow(img)
plt.axis('off')
plt.show()
```

**Creating a user-defined function to recognize text from an image, where the image is opened using `PIL.Image`, and `pytesseract.image_to_string` is used to extract the text from the image.**

```python
# Defining the User-defined function to Recognize Text from the Image
def rec_txt(filename):
    text = pytesseract.image_to_string(Image.open(filename))
    return text
```

```python
# Passing the image to the User-defined function and printing the Result
result = rec_txt(filename)
print(result)
```
