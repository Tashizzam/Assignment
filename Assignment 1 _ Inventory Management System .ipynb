```python
products = {
	1 : {'product_name': '5star', 'quantity': 10, 'cost': 5},
	2 : {'product_name': 'Dairy Milk', 'quantity': 10, 'cost': 5}
}

Total = 0
n = 0

def cal(a , b : int) -> int:
	return products[a]['cost'] * b

def check(x , y):
	if (products[x]['quantity'] >= y):
		Total += cal(x, y)
		products[x]['quantity'] -= y
	else:
		print("Not enough stock!!!")

while True:
	print("Enter product id or product name:\n\n")
	p_name = input()
	if int(p_name) in products:
		p_quantity =  int(input(f"Enter quantity of {products [int(p_name)]['product_name']}:\n\n"))
		check(int(p_name) , p_quantity)

	elif any(product['product_name'] == p_name for product in products.values()):
		p_quantity = int(input(f"Enter quantity of {p_name}:\n\n"))
		P_name = list(filter(lambda x: products[x]['product_name'] == p_name, products))[0]
		check(P_name , p_quantity)

	else:
		print("Invalid input!!!\n\n")
		continue

	n = int(input("Enter 0 to continue with the purchase or any other key to view the total bill : \n\n"))
	if n != 0:
		break

print(f"Total bill ammount : {Total}\n\n")
print(products)
```
