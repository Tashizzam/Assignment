# **Assignment 1 : Inventory Management System**

# 1. Importing the Pandas library to organize inventory data in a DataFrame format

**Description: pandas is a Python library that provides data structures and functions needed to work with structured data seamlessly.**

```python
import pandas as pd
```

# 2. Initializing a dictionary to compile the inventory data into a dataset.

**Description: A dictionary is an unordered collection of key-value pairs. Dictionaries are mutable, and keys must be unique. They are used for storing data that is associated with unique keys.**

```python
products = {
	'1' : {'product_name': '5star', 'quantity': 10, 'cost': 5},
	'2' : {'product_name': 'Dairy Milk', 'quantity': 10, 'cost': 5}
}
print("Inventory In Data Frame Format:\n")
df = pd.DataFrame(products)
print(df)
```

# 3. Setting up a global variable `Total` that calculates the overall amount.

**Creating a new integer variable, `n`, which takes input to either continue or end the purchase process.**

**Also a list `purchased_items` that stores purchased items during the shopping process.**

```python
Total = 0
n = 0
purchased_items = []
```

# 4. The `cal` function calculates the total cost by multiplying the product's cost (from the `products` dictionary) by the given quantity `b`.

**Description: function is a block of reusable code that performs a specific task.**

```python
def cal(a , b : int) -> int:
	return products[a]['cost'] * b
```

# 5. The `check` function verifies product stock, updates the total cost if sufficient, and adjusts inventory accordingly.

```python
# Function to check inventory and update stock after a purchase
def check(x, y):
    global Total  # Declare Total as a global variable so we can modify it.
    if products[x]['quantity'] >= y:
        # Calculate and update total
        Total += cal(x, y)
        # Add item to purchased items list
        purchased_items.append({
            'product_name': products[x]['product_name'],
            'quantity': y,
            'cost': products[x]['cost'],
            'total_cost': cal(x, y)
        })
        # Update product quantity in stock
        products[x]['quantity'] -= y
    else:
        print("Not enough stock!!!")
```

# 6. This script takes product input from the user, verifies stock availability, processes purchases, and prints the total bill along with an updated inventory.

```python
print("Inventory Before Updation:\n")
print(df)

# Start purchasing process
while True:
    print("\n\nEnter product id or product name:")
    p_name = input()

    # Check if input is a valid product ID (matching a key in the products dictionary)
    if p_name in products:
        p_quantity = int(input(f"\n\nEnter quantity of {products[p_name]['product_name']}:\n"))
        check(p_name, p_quantity)

    # If the input is a product name, search for the corresponding product ID
    elif any(product['product_name'] == p_name for product in products.values()):
        p_quantity = int(input(f"\n\nEnter quantity of {p_name}:\n"))
        P_name = list(filter(lambda x: products[x]['product_name'] == p_name, products))[0]
        check(P_name, p_quantity)

    else:
        print("Invalid input!!!\n")
        continue

    # Ask if the user wants to continue or view the total bill
    n = int(input("\n\nEnter 0 to continue with the purchase and any other for total bill: \n"))
    if n != 0:
        break

# Print the purchased items in a formal grocery bill format
print("\n------ GROCERY BILL ------\n")
print("{:<20} {:<10} {:<10} {:<10}".format("Product", "Quantity", "Unit Price", "Total Cost"))
print("-" * 50)

# Loop through purchased items and display them
for item in purchased_items:
    print("{:<20} {:<10} {:<10} {:<10}".format(item['product_name'], item['quantity'], item['cost'], item['total_cost']))

print("-" * 50)
print(f"{'TOTAL':<40} {Total:<10}")
print("\n")

# Display updated inventory after purchases
print("Inventory After Updation:\n")
print(pd.DataFrame(products))
```
